Questão 1:

Assinale a **alternativa correta** acerca da modelagem conceitual do processo de ETL:

a.O esquema conceitual não deve englobar a evolução das regras de negócio.<br>
b.Não é necessário que o desenvolvedor adote um modelo que represente um workflow de ETL, mas sim que ele realize a construção imediata de todo este processo em um código fonte com bons comentários.<br>
c.A melhor forma de se modelar o processo de ETL é após o fim de sua implementação, permitindo que o desenvolvedor tenha uma melhor visão do que foi criado.<br>
d.Uma boa modelagem sempre é complexa e sempre tem múltiplos fluxos.<br>
<font color='red'>e.A modelagem conceitual do processo de ETL baseada em operadores contribui para facilitar a análise de impacto e a refatoração do processo em casos de manutenção e evolução do sistema.</font><br>

Questão 2:

Assinale a **alternativa correta** acerca dos níveis de abstração de um modelo:

a.O nível lógico possui o nível mais alto de abstração, facilitando o entendimento do processo.<br>
b.O nível lógico está próximo à forma na qual os usuários pensam, mas longe dos detalhes de implementação.<br>
c.O nível conceitual possui o nível mais baixo de abstração, incorporando detalhes de implementação e otimização.<br>
d.O nível conceitual possui características próximas da implementação, apesar de já estar mais perto da forma na qual o usuário pensa.<br>
<font color='red'>e.O nível físico possui o nível mais baixo de abstração, incorporando detalhes de implementação e otimização.</font><br>

Nível Conceitual => Nível Alto de Abstração => complementa a análise de requisitos, facilitando o entendimento do processo

Nível Lógico => Nível Médio de Abstração => descreve os detalhes técnicos das tarefas envolvidas

Nível Físico => Nível Baixo de Abstração => incorpora aspectos de implementação e otimização

Questão 3:

Assinale a **alternativa correta** sobre operadores de manipulação de dados do Modelo Intuitive:

<font color='red'>a.O operador Diff representa a ação de calcular as diferenças entre os itens de dados de dois conjuntos.</font><br>
b.O operador Filter é usado para juntar os itens de dados de dois conjuntos fornecidos.=>Join<br>
c.O operador Split é usado para combinar dois itens de dados de dois conjuntos fornecidos.=>Union/Intersect<br>
d.O operador Intersect realiza a ordenação dos itens de um conjunto de dados, em ordem crescente ou decrescente.=>Sort<br>

Operadores de Manipulação de Dados: Representam operações de transformação e de limpeza dos dados

Filter, Sort, Update, Union, Intersect, Diff, Split, Join, Copy
